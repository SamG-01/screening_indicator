```python
import numpy as np
from scipy.optimize import curve_fit

from matplotlib import pyplot as plt, style
from ipywidgets import interact, IntSlider, FloatSlider

import chugunov_indicator as chug
```

```python
style.use("default")
```

We make the assumption that our intercept function is separable, i.e.
$$
    C(\bar{A}, \overline{Z^2}, Z_1, Z_2) = C_0 + C_1(\overline{Z^2}) + C_2(\bar{A}) + C_3(Z_1, Z_2),
$$
for some constant $C_0$ and functions $C_1, C_2, C_3$, and that screening is skippable if
$$
    C \leq 3 \log_{10} T - \log_{10} \rho =: y_0.
$$

Based on the results of the previous notebook, we assume that $C$ is independent of $\bar{Z}, A_1, A_2$.

To test this, we need to examine slices of $C$ on a 4D grid for $\bar{A}, \overline{Z^2}, Z_1, Z_2$.

```python
T_ = np.logspace(7, 9.35, 100)
D_ = np.logspace(-4, 8, 100)

T, D = np.meshgrid(T_, D_)
```

```python
minmaxes = {
    "abar": FloatSlider(value=1, min=1, max=75),
    "z2bar": FloatSlider(value=1, min=1, max=1000),
    "z1": IntSlider(value=1, min=1, max=25),
    "z2": IntSlider(value=1, min=1, max=25)
}

linspaces = {param: np.linspace(slider.min, slider.max, 100) for param, slider in minmaxes.items()}
linspaces["z2bar"] = np.logspace(0, 3, 100)
```

```python
def configuration_space(x: str) -> tuple[dict, dict]:
    minmaxes_ = {**minmaxes}
    del minmaxes_[x]

    X = {x: linspaces[x][...]}
    return X, minmaxes_

def C(X, params) -> np.ndarray:
    X = {x: v[..., None, None, None] for x, v in X.items()}
    return chug.intercept_from_vars(T, D, **X, **params).squeeze()
```

First, we make a plot of $C_1(\overline{Z^2})$.

```python
%matplotlib widget

fig1, ax1 = plt.subplots(constrained_layout=True)

X1, minmaxes1 = configuration_space("z2bar")

ax1.set_title("$C_1$ vs $\\overline{Z^2}$")
ax1.set_xscale("log")

mins1 = {k: v.min for k, v in minmaxes1.items()}
z2bar = linspaces["z2bar"]
log_z2bar = np.log10(z2bar)

C1 = C(X1, mins1)
C1 -= (y0 := C1[0])

line1, = ax1.plot(z2bar, C1, linestyle="dotted")

@interact(**minmaxes1)
def update1(**kwargs) -> None:
    C1 = C(X1, kwargs)
    C1 -= C1[0]
    line1.set_ydata(C1)

    grad = np.gradient(C1, log_z2bar)
    m, dm = grad.mean(), grad.std()
    line1.set_label(f"$a_1 = {m:.3f} \\pm {dm:.3g}$")

    ax1.legend()

#fig1.show()
```

```python
fig1.savefig("C1.svg")
```

As revealed above, we have $C_1 \propto \log_{10} \overline{Z^2}$, with the proportionality constant $m$ being between 1 and 1.10, depending on the values of $\bar{A}, Z_1, Z_2$. We can then average this slope over a grid of this data to find a reasonable slope to use for all cases.

```python
_z2bar, _abar, _z1, _z2, _T, _D = np.meshgrid(
    z2bar[::10], linspaces["abar"][::10], linspaces["z1"][::10], linspaces["z2"][::10],
    T_, D_, sparse=True
)

_C = chug.analytic.intercept_from_vars(_T, _D, abar=_abar, z2bar=_z2bar, z1=_z1, z2=_z2)
```

```python
grad = np.gradient(_C, log_z2bar[::10], axis=1)
M = grad.mean(axis=0)
(m1 := M.mean()), M.std()
```

As such, we can safely set $C_1(\overline{Z^2}) = 1.03 \log_{10} \overline{Z^2}$.

From there, we make the graph for $\bar{A}$.

```python
fig2, ax2 = plt.subplots(constrained_layout=True)

X2, minmaxes2 = configuration_space("abar")

ax2.set_title("$C_2$ vs $\\bar{A}$")

mins2 = {k: v.min for k, v in minmaxes2.items()}
abar = linspaces["abar"]

C2 = C(X2, mins2)
C2 -= (_C0 := C2[0])

curve2 = ax2.scatter(abar, C2, color="red", s=5, label=r"$y = C(\bar{A}, 1; 1, 1)$")

predict2, = ax2.plot(
    abar, 10.4*(abar**(-0.0207) - abar**0.0207),
    linestyle="--", color="green",
    label=r"$y = 10.4(\bar{A}^{-0.0207} - \bar{A}^{0.0207})$"
)
ax2.legend()

@interact(**minmaxes2)
def update2(**kwargs) -> None:
    C2 = C(X2, kwargs)
    C2 -= _C0
    curve2.set_offsets(np.vstack((abar, C2)).T)
    predict2.set_ydata(C2 + C2[0])

#fig2.show()
```

```python
fig2.savefig("C2.svg")
```

Based on the shape of this graph, we anticipate a function of the form
$$
    C_2(\bar{A}) = a_2 (\bar{A}^{-k_2} - \bar{A}^{k_2}),
$$
for positive $a_2, k_2$ and $k_2$ small. We could allow for different coefficients and exponents on each of the terms, but the fit should be effective enough without that.

Since variations in $\overline{Z^2}, Z_1, Z_2$ only change the height of the graph, we don't have to fit on a grid this time.

```python
def C2_func(abar: np.ndarray, a2: float, k2: float) -> np.ndarray:
    return a2 * (abar**(-k2) - abar**(k2))
```

```python
popt, pcov = curve_fit(C2_func, abar, C2, p0=[8, 0.33])
popt, pcov
```

```python
a2, k2 = popt
```

A plot of $C_2$ with these parameters is included in the graph above.

Finally, while $C_3(Z_1, Z_2)$ is symmetric in $Z_1, Z_2$, we don't anticipate further separability, as `chugunov_2009` has a significant dependence on the product $Z_1 Z_2$.

```python
fig3, ax3 = plt.subplots(constrained_layout=True)

X3, minmaxes3 = configuration_space("z1")

ax3.set_title("$C_3$ vs $Z_1$")

mins3 = {k: v.min for k, v in minmaxes3.items()}
z1 = linspaces["z1"]
z2 = mins3["z2"]

C3 = C(X3, mins3)
y0 = C3[0]

a3, k3 = 7.73, 0.026

curve3, = ax3.plot(z1, C3, linestyle="dotted", color="red", label=r"$y = C_3(1, 1; Z_1, 4)$")
predict3, = ax3.plot(z1, y0 + a3*((z1**k3 + z2**k3)**2 - (1 + z2**k3)**2), linestyle="--", color="green")

predict3.set_label(r"$y = 7.73[(Z_1^{0.026} + Z_2^{0.026})^2 - 4]$")
ax3.legend()

@interact(**minmaxes3)
def update3(**kwargs) -> None:
    C3 = C(X3, kwargs)
    curve3.set_ydata(C3)

    z2 = kwargs["z2"]
    _C3 = a3*((z1**k3 + z2**k3)**2 - (1 + z2**k3)**2)
    predict3.set_ydata(_C3 + C3[0])
    ax3.autoscale_view()

#fig3.show()
```

```python
fig3.savefig("C3.svg")
```

Lastly, we make the ansatz
$$
    C_3(Z_1, Z_2) = a_3 [(Z_1^{k_3} + Z_2^{k_3})^2 - 4]
$$
for positive $a_3, k_3$ and $k_3$ small. The exponent has to be $k_3$ for both by symmetry, and the term in parentheses is squared so that $C_3$ has a term proportional to $(Z_1 Z_2)^{k_3}$ in it.

```python
def C3_func(Z: tuple[np.ndarray, np.ndarray], a3: float, k3: float) -> np.ndarray:
    return a3 * ((Z[0]**k3 + Z[1]**k3)**2 - 4)
```

```python
_z1, _z2, _T, _D = np.meshgrid(
    __z1 := linspaces["z1"][::5], __z2 := linspaces["z2"][::5],
    T_, D_, sparse=True
)

_C = chug.analytic.intercept_from_vars(_T, _D, abar=1, z2bar=1, z1=_z1, z2=_z2)
_C -= chug.analytic.intercept_from_vars(_T, _D, abar=1, z2bar=1, z1=1, z2=1)
```

```python
_z1_, _z2_ = np.meshgrid(__z1, __z2)
_z1_, _z2_ = _z1_.ravel(), _z2_.ravel()
```

```python
popt, pcov = curve_fit(C3_func, (_z1_, _z2_), _C.ravel(), p0=[7.65, 0.0262])
popt, pcov
```

```python
a3, k3 = popt
```

Finally, since $C_1(1) = C_2(1) = C_3(1, 1) = 0$, we get $C_0 = C(1, 1, 1, 1)$.

```python
C0 = chug.analytic.intercept_from_vars(_T, _D, abar=1, z2bar=1, z1=1, z2=1)[0, 0]
C0
```

With that, we can put together the final function for predicting whether screening can be skipped:
$$
    C(\bar{A}, \overline{Z^2}, Z_1, Z_2) \approx 20.79 + 1.03 \log_{10} \overline{Z^2} + 10.42(\bar{A}^{-0.02} - \bar{A}^{0.02}) + 7.73[(Z_1^{0.026} + Z_2^{0.026})^2 - 4].
$$
Absorbing all constant terms into $C_0$ simplifies this to
$$
    C(\bar{A}, \overline{Z^2}, Z_1, Z_2) \approx -10.132 + 1.03 \log_{10} \overline{Z^2} + 10.42(\bar{A}^{-0.02} - \bar{A}^{0.02}) + 7.73(Z_1^{0.026} + Z_2^{0.026})^2.
$$
